```python
#for data sourcing

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('bank.csv')
```

<h1 style="color:Turquoise;">Data description</h1>
The data is related with direct marketing campaigns of a Portuguese banking institution. The marketing campaigns were based on phone calls. Often, more than one contact to the same client was required, in order to access if the product (bank term deposit) would be ('yes') or not ('no') subscribed. 
bank.csv with 10% of the examples and 17 inputs, randomly selected from 3 (older version of this dataset with less inputs). 
The smallest datasets are provided to test more computationally demanding machine learning algorithms. 

The classification goal is to predict if the client will subscribe (yes/no) a term deposit (variable y).

<h1 style="color:orange;">Feature Encoding</h1>

```python
cleandata=data
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cleandata['job'] = le.fit_transform(cleandata['job'])
cleandata['marital'] = le.fit_transform(cleandata['marital'])
cleandata['education'] = le.fit_transform(cleandata['education'])
cleandata['default'] = le.fit_transform(cleandata['default'])
cleandata['housing'] = le.fit_transform(cleandata['housing'])
cleandata['loan'] = le.fit_transform(cleandata['loan'])
cleandata['contact'] = le.fit_transform(cleandata['contact'])
cleandata['month'] = le.fit_transform(cleandata['month'])
cleandata['poutcome'] = le.fit_transform(cleandata['poutcome'])
cleandata['y'] = le.fit_transform(cleandata['y'])
cleandata.info()
```

```python
cleandata.corr()
cleandata.corr()['y'].sort_values(ascending=False)
```

<h1 style="color:lime;">Feature V/S Target</h1>

```python
features = cleandata.drop('y',axis=1)
target = cleandata['y']
```

```python
features.head(16)
```

<h1 style="color:#FD5B78;">Decision Tree Classifier</h1>

<h1 style="color:fuchsia;">Decision Tree Classifier Training</h1>

```python
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test= train_test_split(features, target, test_size= 0.25, random_state=0)
```

```python
#Fitting Decision Tree classifier to the training set
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion='entropy', random_state=0) 
model.fit(x_train, y_train)
```

```python
#checking my model fitting
model.score(x_train, y_train)
```

<h1 style="color:fuchsia;">Decision Tree Classifier Testing</h1>

```python
#Predicting the test set result
#model.predict([[26,3,1,2,0,543,0,0,0,30,4,169,3,-1,0,3]])
y_pred= model.predict(x_test)
```

<h1 style="color:#FF007C;">Decision Tree Classifier Score</h1>

```python
#model gives a score of 87 out of 100 to the factual entries
model.score(x_test, y_test)
```

<h1 style="color:#FF007C;">Passed with Flying colors - 87% 🎉</h1>

```python
#Creating the Confusion matrix  
from sklearn.metrics import confusion_matrix  
cm= confusion_matrix(y_test, y_pred)  
cm
#71 + 73 = 144 incorrect predictions and 922 + 65 = 987 correct predictions
```
